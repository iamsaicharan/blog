```python
import pandas as pd

data = pd.read_csv('titanic.csv')

data.head()
```

```python
joint_prob = data.groupby(['Survived', 'Sex']).size().reset_index(name='counts')
joint_prob['prob'] = joint_prob['counts'] / data.shape[0]
```

```python
marginal_prob = data.groupby('Survived').size().reset_index(name='counts')
marginal_prob['prob'] = marginal_prob['counts'] / data.shape[0]
```

```python
cond_prob = pd.merge(joint_prob, marginal_prob, on='Survived')
cond_prob['cond_prob'] = cond_prob['prob_x'] / cond_prob['prob_y']
cond_prob = cond_prob.drop(['counts_y', 'prob_y'], axis=1)
cond_prob = cond_prob.rename(columns={'Survived': 'Survived', 'Sex': 'Gender', 'counts_x': 'Count', 'prob_x': 'Probability', 'cond_prob': 'Conditional Probability'})
```
